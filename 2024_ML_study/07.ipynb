### array operation
** array 간 계산 ! <br>
매우 중요한 개념 : **broadcasting** <br>
이게 뭐냐 => shape이 다른 배열 간 연산을 자동으로 지원하는 기능 

```python
import numpy as np 
```

```python
# 배열 생성
test_a = np.array([[1, 2, 3], [4, 5, 6]], float)

# 배열 출력
print(test_a)

# Matrix + Matrix 연산
print(test_a + test_a)

# Matrix - Matrix 연산
print(test_a - test_a)

# Matrix * Matrix 연산 (각 elements를 곱함)
print(test_a * test_a)

# 배열 간 shape 이 같다면 ! 
```

```python
# 두 배열 생성
# test_a는 1부터 6까지의 숫자로 구성된 2x3 배열
# test_b는 7부터 12까지의 숫자로 구성된 3x2 배열
test_a = np.arange(1, 7).reshape(2, 3)
test_b = np.arange(7, 13).reshape(3, 2)

# 배열 출력
print("test_a:")
print(test_a)
print("\ntest_b:")
print(test_b)

# Dot Product (행렬 곱)
# Matrix A의 각 행과 Matrix B의 각 열을 곱한 후 합산한 값을 결과로 하는 행렬을 생성
result = test_a.dot(test_b) # 이게 중요 

# 결과 출력
print("\nDot Product (행렬 곱) 결과:")
print(result)
```

```python
# transpose 
print (test_a.T) 
print (test_a.T.dot(test_a))
```

```python
### broadcasting 예시

# 2x3 행렬을 생성
test_matrix = np.array([[1, 2, 3], [4, 5, 6]], float)

# 스칼라 값 3을 설정
scalar = 3

# Broadcasting을 이용한 덧셈 연산
# 스칼라 값이 행렬의 각 요소에 더해짐
result = test_matrix + scalar

# 결과 출력
print("test_matrix + scalar:")
print(result)
```
