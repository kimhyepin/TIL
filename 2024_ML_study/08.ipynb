```python
import numpy as np 
```

```python
# 0부터 9까지의 값을 가진 배열 생성
a = np.arange(10)
print("Array a:")
print(a)

# np.any() 함수 사용 예제
# 배열 a에서 5보다 큰 값이 하나라도 있는지 확인 
print("\nnp.any(a > 5):")
print(np.any(a > 5)) 

# 배열 a에서 0보다 작은 값이 하나라도 있는지 확인
print("np.any(a < 0):")
print(np.any(a < 0)) 

# np.all() 함수 사용 예제
# 배열 a의 모든 값이 5보다 큰지 확인
print("\nnp.all(a > 5):")
print(np.all(a > 5))  

# 배열 a의 모든 값이 10보다 작은지 확인
print("np.all(a < 10):")
print(np.all(a < 10))  

print("\n5보다 큰 숫자가 있는지 list형태로:")
print (a>5) 
```

```python
# 두 배열 생성
test_a = np.array([1, 3, 0], float)
test_b = np.array([5, 2, 1], float)

# 각 요소에 대한 비교 연산
print("test_a > test_b:")
print(test_a > test_b)  # 각각의 요소를 비교하여 Boolean 배열 반환

# 두 배열의 요소가 같은지 비교
print("\ntest_a == test_b:")
print(test_a == test_b)  # 각각의 요소를 비교하여 Boolean 배열 반환

# 배열 내 하나라도 조건을 만족하는지 확인
print("\n(test_a > test_b).any():")
print((test_a > test_b).any())  # 하나라도 True가 있으면 True 반환
```

```python
# 배열 a 생성
a = np.array([1, 3, 0], float)

# 논리적 AND 연산
print("np.logical_and(a > 0, a < 3):")
print(np.logical_and(a > 0, a < 3))  # AND 조건: 두 조건이 모두 참인 경우만 True

# 배열 b 생성
b = np.array([True, False, True], bool)

# 논리적 NOT 연산
print("\nnp.logical_not(b):")
print(np.logical_not(b))  # NOT 조건: True를 False로, False를 True로 변환

# 배열 c 생성
c = np.array([False, True, False], bool)

# 논리적 OR 연산
print("\nnp.logical_or(b, c):")
print(np.logical_or(b, c))  # OR 조건: 하나라도 참인 경우 모두 True, list 형태 반환 
```

np.where 

```python
# 배열 a 생성
a = np.array([1, 3, 0], float)

# np.where를 사용하여 조건에 따라 값을 선택
# a > 0 조건을 만족하는 경우 3, 그렇지 않으면 2를 선택
result = np.where(a > 0, 3, 2) # 참이면 3, 거짓이면 2 // 엑셀과 똑같음 

# 결과 출력
print("np.where(a > 0, 3, 2):")

# 0부터 9까지의 값을 가진 배열 생성
a = np.arange(10)

# np.where를 사용하여 a > 5 조건을 만족하는 요소의 인덱스 반환
indices = np.where(a > 5)

# 인덱스 출력
print("\nnp.where(a > 5):")
print(indices)

# NaN 및 Inf 값을 포함한 배열 생성
a = np.array([1, np.NaN, np.Inf], float)

# np.isnan을 사용하여 NaN인 요소를 찾아 Boolean 배열로 반환
nan_check = np.isnan(a)

# 결과 출력
print("\nnp.isnan(a):")
print(nan_check)

# np.isfinite를 사용하여 유한한 숫자인지 확인하는 Boolean 배열 반환
finite_check = np.isfinite(a)

# 결과 출력
print("\nnp.isfinite(a):")
print(finite_check)
```

argmax & argmin 

```python
a= np.array ([1, 2, 4, 5, 8, 78, 23, 31])
np.argmax(a), np.argmin(a)
# 78과 1의 인덱스가 반환됨 
```

```python
# axis 기반의 반환 

a = np.array([[1,2,4,7],[9,88,6,45],[9,76,3,4]])
np.argmax(a, axis=1), np.argmin(a, axis=0)
```
