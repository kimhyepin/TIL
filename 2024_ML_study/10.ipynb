### pandas 
구조화된 데이터의 처리를 지원하는 python 라이브러리 <br> 즉, python계의 excel <br><br>

**series** <br>
Pandas에서 Series는 DataFrame의 한 열을 나타내며, 여러 Series가 결합되어 DataFrame을 형성한다. <br>
Series는 데이터와 인덱스, 데이터 타입, 이름 등을 포함하는 객체이다. <br>
인덱스는 데이터의 구분을 위한 이름으로, 데이터를 합칠 때 중요한 역할을 한다. <br>
Series는 numpy 배열의 서브클래스이며, 쉽게 데이터를 핸들링할 수 있는 기능을 제공한다. <br>
Series의 인덱스는 중복을 허용하지만, 일반적으로는 중복 사용이 권장되지 않는다. <br>

```python
import pandas as pd
```

```python
data = [1, 2, 3, 4, 5]
index = ['a', 'b', 'c', 'd', 'e']
series = pd.Series(data, index=index)
print(series)
```
