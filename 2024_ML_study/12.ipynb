```python
# 'account' 컬럼을 Series로 선택
account_series = df["account"]

# 처음 3개의 데이터를 선택
print(account_series[:3])

# 특정 인덱스(0, 1, 2)를 선택하여 출력
print(account_series[[0, 1, 2]]) ## 앞과 같은 결과 

# Boolean index를 사용하여 조건에 맞는 데이터 선택 (250000보다 작은 값)
print(account_series[account_series < 250000]) 
a = account_series < 250000
print (a) # boolean series  
# 이 a라는 불린 시리즈 이용해서 원래 시리즈에서 True 인 항목만 필터링하여 출력
```
